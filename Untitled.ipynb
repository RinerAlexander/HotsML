```python
import requests
import json
from pprint import pprint
url = "http://hotsapi.net/api/v1/replays?min_id=30688945&existing=true&with_players=true"
raw_data=requests.get(url).json()
pprint(raw_data)
```

```python
game=raw_data[0]
first_team=[]
second_team=[]
for player in game["players"]:
    name=player["hero"]

    if player["team"]==0:
        first_team.append(name)
        first_team_won=player["winner"]
    else:
        second_team.append(name)
```

```python
print(first_team)
print(first_team_won)
print(second_team)
```
