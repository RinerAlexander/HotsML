```python
import requests
import json
from pprint import pprint
```

```python
url = "http://hotsapi.net/api/v1/replays?min_id=30000000&existing=true&with_players=true"
```

```python
raw_data=requests.get(url).json()
```

```python
pprint(raw_data[1])
```

```python
data=[]
for game in raw_data:
    first_team=[]
    second_team=[]
    for player in game["players"]:
        if game["players"][player]["team"]==0:
            first_team.append(game["players"][player]["hero"])
        else:
            second_team.append(game["players"][player]["hero"])
    game_data={
        "first":first_team,
        "second":second_team,
        "map":game["game_map"],
        "type":game["game_type"],
        "first_won":game["teams"][0]["winner"]
    }
    data.append(game_data)
```

```python
pprint(data)
```
