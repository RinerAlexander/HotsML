```python
import requests
import json
from pprint import pprint
import re
```

```python
url = "http://hotsapi.net/api/v1/replays?min_id=30400000&existing=true&with_players=true"
raw_data=requests.get(url).json()
```

```python
pprint(raw_data)
```

```python
game=raw_data[1]
first_team=[]
second_team=[]
for player in game["players"]:
        if game["players"][player]["team"]==0:
            first_team.append(game["players"][player]["hero"])
        else:
            second_team.append(game["players"][player]["hero"])
print(first_team)
print(second_team)
```

```python
data=[]
for game in raw_data:
    first_team=[]
    second_team=[]
    for player in game["players"]:
        name=game["players"][player]["hero"]
        
        #Newer hero's don't have names but we are able to pull the name from the 
        #talents
        if name == None:
            name = game["players"][player]["talents"]["1"]
            name=re.sub( r"([A-Z])", r" \1", name).split()[0]
            
        if game["players"][player]["team"]==0:
            first_team.append(name)
        else:
            second_team.append(name)
            
    game_data={
        "first":first_team,
        "second":second_team,
        "map":game["game_map"],
        "type":game["game_type"],
        "first_won":game["teams"][0]["winner"]
    }
    data.append(game_data)
```

```python
pprint(data)
```

```python
id_min=30400000
data=[]
more_games=True
while more_games:
    url = f"http://hotsapi.net/api/v1/replays?min_id={id_min}&existing=true&with_players=true"
    raw_data=requests.get(url).json()
    
    if raw_data:
        
        for game in raw_data:
            first_team=[]
            second_team=[]
            for player in game["players"]:
                name=game["players"][player]["hero"]

                #Newer hero's don't have names but we are able to pull the name from the 
                #talents
                if name == None:
                    name = game["players"][player]["talents"]["1"]
                    name=re.sub( r"([A-Z])", r" \1", name).split()[0]

                if game["players"][player]["team"]==0:
                    first_team.append(name)
                else:
                    second_team.append(name)

            game_data={
                "first":first_team,
                "second":second_team,
                "map":game["game_map"],
                "type":game["game_type"],
                "first_won":game["teams"][0]["winner"]
            }
            data.append(game_data)
    else:
        more_games=False
```
